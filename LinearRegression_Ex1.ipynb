```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/Users/huy/Documents/MÁY HỌC THỐNG KÊ/data/Income.csv')
y = df[['Expenditure']]
x = df[['Income']]
```

```python
model = LinearRegression().fit(x,y)
```

```python
# lấy kết quả
intercept = model.intercept_
coefs = model.coef_
print(coefs[0][0])
R_2 = model.score(x,y)
R_2

# visualization
plt.scatter(x, y)
plt.plot(x, intercept + coefs*x, color='orange', lw=2, linestyle='--')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.show()
```

```python
# prediction
predited_values = intercept + coefs*x
plt.plot(x,y, color="blue", label="Actual", linestyle="--")
plt.plot(x, predited_values, color="red", label="Predicted", linestyle="--", marker="x")
plt.legend()
plt.show()
```

```python
# future
orthers_x = np.array([28, 35, 46, 52]).reshape(-1,1)
pre_orther_values = model.predict(orthers_x)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 13
```

```python
# load data
df = pd.read_excel("/Users/huy/Documents/MÁY HỌC THỐNG KÊ/data/Sales.xlsx")
x = df[["Price", "Ads_Cost"]]
y = df[["Sales_Volume"]]
x,y = np.array(x), np.array(y)
x = sm.add_constant(x)
```

```python
# fit model
model = sm.OLS(y, x)
results = model.fit()
```

```python
# get results
R_2 = results.rsquared
prms = results.params
print(f"R^2: {R_2}")
print(f"Parameters: {prms}")
print(results.summary())
```

## giải thích
y = a + b.price + c.ads_cost

a = 292.5
b = -24.8
c = 77.9

a là hệ số gốc -> nếu giá không đổi và không có đầu tư vào chi phí quảng cáo thì lợi nhuận nhận được là 292.5

b -> nếu b tăng lên 1 đơn vị thì y(lợi nhuận) giảm đi 24.8 đơn vị

c -> nếu c tăng lên 1 đơn vị thì y(lợi nhuận) tăng lên 77.9 đơn vị




```python
# predict
predicted_value = results.predict(x)
plt.plot(df["Sales_Volume"], label="Actual")
plt.plot(predicted_value, label="Predicted")
plt.legend()
plt.show()
```

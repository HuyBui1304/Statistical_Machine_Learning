```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 13
```

```python
# Load data
data = pd.read_csv('/Users/huy/Documents/MÁY HỌC THỐNG KÊ/SampleData/HomePrices.csv')

x1 = data[['Area']]
y1 = data[['Price']]
x2 = data[['Bedrooms']]
y2 = data[['Price']]
x3 = data[['Age']]
y3 = data[['Price']]
x4 = data[['Area', 'Bedrooms']]
y4 = data[['Price']]
x5 = data[['Area', 'Age']]
y5 = data[['Price']]
x6 = data[['Bedrooms', 'Age']]
y6 = data[['Price']]
x7 = data[['Area', 'Bedrooms', 'Age']]
y7 = data[['Price']]

x1, y1 = np.array(x1), np.array(y1)
x1 = sm.add_constant(x1)
x2, y2 = np.array(x2), np.array(y2)
x2 = sm.add_constant(x2)
x3, y3 = np.array(x3), np.array(y3)
x3 = sm.add_constant(x3)
x4, y4 = np.array(x4), np.array(y4)
x4 = sm.add_constant(x4)
x5, y5 = np.array(x5), np.array(y5)
x5 = sm.add_constant(x5)
x6, y6 = np.array(x6), np.array(y6)
x6 = sm.add_constant(x6)
x7, y7 = np.array(x7), np.array(y7)
x7 = sm.add_constant(x7)
```

```python
# fit model
model1 = sm.OLS(y1,x1)
result1 = model1.fit()
model2 = sm.OLS(y2,x2)
result2 = model2.fit()
model3 = sm.OLS(y3,x3)
result3 = model3.fit()
model4 = sm.OLS(y4,x4)
result4 = model4.fit()
model5 = sm.OLS(y5,x5)
result5 = model5.fit()
model6 = sm.OLS(y6,x6)
result6 = model6.fit()
model7 = sm.OLS(y7,x7)
result7 = model7.fit()

```

```python
# get results
def print_regression_results(result):
    """
    Hàm nhận đầu vào là một mô hình hồi quy (result) 
    và in ra các kết quả: R^2, các hệ số ước lượng, và bảng summary.
    """
    R_2 = result.rsquared
    prms = result.params
    print(f"R^2: {R_2}")
    print(f"Parameters:\n{prms}\n")
    print(result.summary())

```

```python
#%% Perform stepwise regression
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
model = LinearRegression()
rfe = RFE(model, n_features_to_select=2)
fit = rfe.fit(x7, y7)

print("Feature ranking: ")
print(fit.ranking_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (10,8)
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.size'] = 25
```

```python
def cost_function(theta):
    return theta**2 - 4*theta + 8
```

```python
def gradient(theta):
    return 2* theta - 4
```

```python
def gradient_descent(starting_theta, learning_rate, eps):
    theta = starting_theta
    theta_history = [theta]
    cost_history = [cost_function(theta)]
    
    while True :
        theta = theta - learning_rate * gradient(theta)
        theta_history.append(theta)
        cost_history.append(cost_function(theta))
        if abs(gradient(theta)) < eps:
            break
    
    return theta_history, cost_history
```

```python
starting_theta = 0
learning_rate = 0.1
eps = 0.0001
```

```python
theta_history, cost_history = gradient_descent(starting_theta, learning_rate, eps)
```

```python
theta_values = np.linspace(-1, 5, 400)
cost_values = cost_function(theta_values)

plt.figure(figsize=(18,10))

plt.subplot(1,2,1)
plt.plot(theta_values, cost_values, label ='$J(\\theta) = \\theta^2 - 4\\theta +8$', color='darkgreen', lw =2)
plt.scatter(theta_history,cost_history,color='darkorange',s = 80)
plt.plot(theta_history,cost_history,color='darkorange', linestyle='--',lw=2)

plt.xlabel('Parameters - $\\theta$')
plt.ylabel('Cost - $J(\\theta)$')
plt.title('Cost Function and Gradient Descent Path')
plt.legend()

plt.subplot(1,2,2)
plt.plot(cost_history,label='Cost History', color='darkgreen', lw = 2)
plt.xlabel('Iteration')
plt.ylabel('Cost - $J(\\theta)$')
plt.title('Cost History during Gradient Descent')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
print(f'The cost function reaches the minium value of {cost_history[-1]} when theta ={theta_history[-1]}')
```
